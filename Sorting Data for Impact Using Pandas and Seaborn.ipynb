### Sorting Data for Impact Using Pandas and Seaborn

This notebook simply shows how to aggregate count data and display it using the pandas and seaborn libraries in Python.

Let's start by importing libraries:

````python
import pandas as pd
import seaborn as sns
````

Now, we'll read sample sales data from a csv file contained in the current directory. This file is a Kaggle dataset from the following link: https://www.kaggle.com/kyanyoga/sample-sales-data. Typically, it would pertinent to do a thorough dataset review, but the purpose of this notebook is demonstrate how to aggregate and sort data using pandas as well as display the data using visuals from seaborn.

````python
df=pd.read_csv('sales_data_sample.csv', encoding= 'unicode_escape') 
````

Let's take an inital look at the number of orders by country. We'll start by aggregating the data by country using ```.groupby()``` from pandas followed by ```.count()``` from pandas, we'll also reset the index using ```.reset_index()```.

````python
aggregated_df=df[['COUNTRY','ORDERNUMBER']].groupby(['COUNTRY']).count().reset_index()
aggregated_df=aggregated_df.rename(columns={'ORDERNUMBER':'ORDER COUNT'})
````

Next, we'll create a simple bar chart that displays the aggregated data. We'll use seaborn's bar plot indicated by ```sns.barplot()```. We'll set the x-axis labels using ```.set_xticklabels()``` and set the title using ```.set_title()```.

````python
ax = sns.barplot(x="COUNTRY", y="ORDER COUNT", data=aggregated_df,palette='mako')
ax.set_xticklabels(ax.get_xticklabels(),rotation=75)
ax.set_title("Number of Orders by Country")
````

This looks disorganized, it's not simple to see a clear trend of which countries have the most orders. Of course, some countires, like the USA, seem to have more orders relative to other countries (i.e. Belgium).

Let's sort the data by the number of orders per country using ```.sort_values()``` in pandas

````python
sorted_df=aggregated_df.sort_values(
    by='ORDER COUNT',ascending=False)
````

Now, we'll display the data using a bar chart.

````python
ax = sns.barplot(x="COUNTRY", y="ORDER COUNT", data=sorted_df,palette='mako')
ax.set_xticklabels(ax.get_xticklabels(),rotation=75)
ax.set_title("Number of Orders by Country")
````

This looks better, we get a clear idea of how the number of orders per country compare to each toher.

Below, we also show the plot as a horiztonal plot which some may find makes the country names easier to read. To do this, we simply reverse the x and y options in ```sns.barplot()```.

````python
ax = sns.barplot(x="ORDER COUNT", y="COUNTRY", data=sorted_df,palette='mako')
ax.set_title("Number of Orders by Country")
````
